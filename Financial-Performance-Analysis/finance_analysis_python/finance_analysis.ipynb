```python
### Changing date to datetime dtype
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
money_cols = ['Net Income', 'Revenue', 'Expenditure', 'Gross Profit']

for col in money_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace('[^0-9.-]', '', regex=True)   # remove EVERYTHING except numbers
        .astype(float)
    )
```

```python
##checking all dtatypes of colmuns is changed or not ['Net Income', 'Revenue', 'Expenditure', 'Gross Profit']
df.info() 
```

### Calulating KPI ###

```python
df.columns
```

```python
###calculating ['Net Income', 'Revenue', 'Expenditure', 'Gross Profit']
total_revenue = df['Revenue'].sum()
total_expenditure = df['Expenditure'].sum()
total_net_income = df['Net Income'].sum()
total_gross_profit = df['Gross Profit'].sum()

profit_margin = total_net_income / total_revenue

print("Total Revenue:", total_revenue)
print("Total Expenditure:", total_expenditure)
print("Total Net Income:", total_net_income)
print("Total Gross Profit:", total_gross_profit)
print("Profit Margin %:", round(profit_margin*100, 2))
```

```python
# Create Month column from Date
df['Month'] = df['Date'].dt.month_name()

# Calculate monthly revenue
monthly_revenue = df.groupby('Month')['Revenue'].sum()

# Correct month order
month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']

monthly_revenue = monthly_revenue.reindex(month_order)

# Print results
print("Monthly Revenue:\n")
print(monthly_revenue)

print("\nHighest month:", monthly_revenue.idxmax(), "-", monthly_revenue.max())
print("Lowest month:", monthly_revenue.idxmin(), "-", monthly_revenue.min())
```

```python

# create screenshots folder automatically (if not exists)
os.makedirs('screenshots', exist_ok=True)

plt.figure(figsize=(10,5))

plt.plot(monthly_revenue.index, monthly_revenue.values, marker='o')

plt.title('Revenue Monthly Trends')
plt.xlabel('Month')
plt.ylabel('Revenue')

plt.xticks(rotation=45)
plt.grid(True)

# save chart as image
plt.savefig('screenshots/monthly_revenue_trend.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python

# ===============================
# NET INCOME MONTHLY TREND
# ===============================
# create screenshots folder automatically (if not exists)
os.makedirs('screenshots', exist_ok=True)

# Create monthly net income
monthly_net_income = df.groupby('Month')['Net Income'].sum()

# correct order
month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']

monthly_net_income = monthly_net_income.reindex(month_order)

# plot
plt.figure(figsize=(8,4))

plt.plot(monthly_net_income.index,
         monthly_net_income.values,
         marker='o')

plt.title('Net Income Monthly Trend')
plt.xlabel('Month')
plt.ylabel('Net Income')

plt.xticks(rotation=45)
plt.grid(True)

# save chart as image
plt.savefig('screenshots/monthly_net_income.png', dpi=300, bbox_inches='tight')

plt.show()



```

```python
# ===============================
# PROFIT MARGIN % BY MONTH
# ===============================

import os
os.makedirs('screenshots', exist_ok=True)

# monthly totals
monthly_revenue = df.groupby('Month')['Revenue'].sum()
monthly_net_income = df.groupby('Month')['Net Income'].sum()

# correct month order
month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']

monthly_revenue = monthly_revenue.reindex(month_order)
monthly_net_income = monthly_net_income.reindex(month_order)

# calculate monthly profit margin %
monthly_profit_margin = (monthly_net_income / monthly_revenue) * 100

# plot
plt.figure(figsize=(10,5))

plt.plot(monthly_profit_margin.index,
         monthly_profit_margin.values,
         marker='o')

plt.title('Profit Margin % by Month')
plt.xlabel('Month')
plt.ylabel('Profit Margin (%)')

plt.xticks(rotation=45)
plt.grid(True)

# save image
plt.savefig('screenshots/profit_margin_trend.png',
            dpi=300,
            bbox_inches='tight')

plt.show()
```

```python
# ===============================
# AVERAGE DEBT-TO-EQUITY RATIO BY MONTH
# ===============================

os.makedirs('screenshots', exist_ok=True)

#  calculate monthly average ratio (NOT sum)
monthly_de_ratio = df.groupby('Month')['Debt-to-Equity Ratio'].mean()

# correct order
month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']

monthly_de_ratio = monthly_de_ratio.reindex(month_order)

# plot
plt.figure(figsize=(10,5))

plt.plot(monthly_de_ratio.index,
         monthly_de_ratio.values,
         marker='o')

plt.title('Debt-to-Equity Ratio by Month')
plt.xlabel('Month')
plt.ylabel('Debt-to-Equity Ratio')

plt.xticks(rotation=45)
plt.grid(True)

# save image
plt.savefig('screenshots/debt_to_equity_trend.png',
            dpi=300,
            bbox_inches='tight')

plt.show()
```

```python
##Revenue vs Expenditure comparison chart
# =========================================
# REVENUE vs EXPENDITURE COMPARISON
# =========================================

import os
import matplotlib.pyplot as plt

os.makedirs('screenshots', exist_ok=True)

#1ï¸âƒ£ monthly totals
monthly_revenue = df.groupby('Month')['Revenue'].sum()
monthly_expenditure = df.groupby('Month')['Expenditure'].sum()

# correct order
month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']

monthly_revenue = monthly_revenue.reindex(month_order)
monthly_expenditure = monthly_expenditure.reindex(month_order)

#  plot both lines
plt.figure(figsize=(10,5))

plt.plot(monthly_revenue.index,
         monthly_revenue.values,
         marker='o',
         label='Revenue')

plt.plot(monthly_expenditure.index,
         monthly_expenditure.values,
         marker='o',
         label='Expenditure')

plt.title('Revenue vs Expenditure Trend')
plt.xlabel('Month')
plt.ylabel('Amount')

plt.legend()   # â­ show labels
plt.xticks(rotation=45)
plt.grid(True)

#save image
plt.savefig('screenshots/revenue_vs_expenditure.png',
            dpi=300,
            bbox_inches='tight')

plt.show()
```

```python
# =========================================
# SAVE INSIGHTS TO TEXT FILE
# =========================================

import os

os.makedirs('reports', exist_ok=True)

report_path = os.path.join('reports', 'financial_insights_report.txt')

with open(report_path, 'w', encoding='utf-8') as f:

    f.write("FINANCIAL PERFORMANCE INSIGHTS REPORT\n")
    f.write("=====================================\n\n")

    f.write("REVENUE\n")
    f.write(f"Highest Month : {monthly_revenue.idxmax()} ({monthly_revenue.max():,.0f})\n")
    f.write(f"Lowest Month  : {monthly_revenue.idxmin()} ({monthly_revenue.min():,.0f})\n\n")

    f.write("EXPENDITURE\n")
    f.write(f"Highest Month : {monthly_expenditure.idxmax()} ({monthly_expenditure.max():,.0f})\n")
    f.write(f"Lowest Month  : {monthly_expenditure.idxmin()} ({monthly_expenditure.min():,.0f})\n\n")

    f.write("PROFIT MARGIN\n")
    f.write(f"Best Month  : {monthly_profit_margin.idxmax()} ({monthly_profit_margin.max():.2f}%)\n")
    f.write(f"Worst Month : {monthly_profit_margin.idxmin()} ({monthly_profit_margin.min():.2f}%)\n\n")

    f.write("DEBT RISK\n")
    f.write(f"Highest Risk : {monthly_de_ratio.idxmax()} ({monthly_de_ratio.max():.2f})\n")
    f.write(f"Lowest Risk  : {monthly_de_ratio.idxmin()} ({monthly_de_ratio.min():.2f})\n")

print("Insights report saved successfully!")
```
